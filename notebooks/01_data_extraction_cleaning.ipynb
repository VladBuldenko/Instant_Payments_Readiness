```python
import pandas as pd
import numpy as np
```

### Load Raw Bundesbank XLSX File
**Purpose:** Load the original Bundesbank payment statistics file, inspect available sheets, and extract Table 3 and Table 4 for further cleaning.

```python
file_path = "../data/raw/I.Payments_statistics_810262.xlsx"

xlsx = pd.ExcelFile(file_path)

# Load the Table 3, Table 4 and convert them in the .csv format
table_3 = pd.read_excel(xlsx, sheet_name="Table 3")
table4 = pd.read_excel(xlsx, sheet_name="Table 4")

# Save cleaned CSVs to data/interim/
table_3.to_csv("../data/interim/table_3_transactions.csv", index=False, encoding="utf-8")
table4.to_csv("../data/interim/table_4_values.csv", index=False, encoding="utf-8")
```

**Result:** The Bundesbank workbook was loaded successfully, both Table 3 and Table 4 were extracted, and their cleaned CSV versions were saved to `data/interim/` for further processing.
